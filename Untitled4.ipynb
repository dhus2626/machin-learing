```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 8, 13)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.xlabel('X amount')
plt.ylabel('Y amount')
plt.title('Graph of $y = x^2$')

plt.plot(x, y)
plt.show()
```

```python
x1 = np.linspace(-np.pi, np.pi, 143)
x2 = np.linspace(-np.pi, np.pi, 143)

y1 = np.cos(x1)
y2 = np.sin(x2)
```

```python
plt.plot(x1, y1, x2, y2)
plt.show()
```

Figure Size and dpi

```python
plt.figure(figsize = (15, 8), dpi = 120)

plt.plot(x1, y1)
plt.plot(x2, y2)

plt.show()
```

```python
plt.figure(figsize = (15, 8), dpi = 120)

plt.plot(x1, y1, 'r-.')
plt.plot(x2, y2, 'b--')

plt.show()
```

```python
plt.figure(figsize = (15, 8), dpi = 120)

plt.plot(x1, y1, color = 'black')
plt.plot(x2, y2, color = 'red')

plt.show()
```

plt.figure(figsize = (15, 8), dpi = 120)

plt.plot(x1, y1, color = 'yellow', lw = 4.5)
plt.plot(x2, y2, color = 'green', lw = 4.5)

plt.show()

```python
plt.figure(figsize = (15, 8), dpi = 120)

plt.plot(x1, y1, color = 'purple', lw = 4.5, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 4.5, ls = '-.')

plt.show()
```
